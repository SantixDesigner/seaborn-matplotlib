```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd 
import numpy as np
```

<h1>Main</h1>

```python
df = sns.load_dataset('taxis')
data = df.pivot_table(values=['tip'], index=['color'], aggfunc='sum').reset_index()
data
```

<h2>Cálculo de propina promedio según el color del coche</h2>

```python
sns.barplot(data=df, x='color', y='tip', hue='color')
```

<h2>Cálculo de la propina total</h2>

```python
plt.pie(x=data['tip'], labels=['Green','Yellow'], colors=['Green', 'Yellow'], autopct='%0.1f%%')
plt.show()
```

```python
propina = df[df['passengers'] != 0].pivot_table(values='total', index=['passengers'], aggfunc='count')
propina
```

<h2>Cálculo exclusivo de la cantidad de pasajeros en total</h2>

```python
sns.barplot(data=propina, x='passengers', y='total')
```

<h2>Pairplot general</h2>

```python
sns.pairplot(data=df, hue='color')
```
